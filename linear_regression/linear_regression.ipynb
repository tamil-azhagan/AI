```python
#linear regression
import pandas as pd
import math
df=pd.read_csv(r"C:\Users\god\Desktop\AI_lab\linear_reg.csv")
df
```

```python
#x_mean,y_mean
x_mean=df["x"].mean()
y_mean=df["y"].mean()

print("x mean: ",x_mean)
print("y mean: ",y_mean)
```

```python
#x-xbar
list1=[]
for i in df["x"]:
    list1.append(i-x_mean)
df["x-xbar"]=list1
df
    
```

```python
#y-y_bar
list2=[]
for i in df["y"]:
    list2.append(i-y_mean)
df["y-ybar"]=list2
df
```

```python
#(x-xbar)(y-ybar)
list3=[]
for i in range(len(df["x-xbar"])):
    list3.append(df["x-xbar"][i]*df["y-ybar"][i])
df["xbar_ybar"]=list3
df
```

```python
#xbar.sq
list4=[]
for i in df["x-xbar"]:
    list4.append(i*i)
df["x-xbar_sq"]=list4
df
```

```python
#to calculate b1
b1=df["xbar_ybar"].sum()/df["x-xbar_sq"].sum()
print("The value of b1 is: ",b1)
```

```python
#to calculate y^
#y^=b0+b1x
#initially to calculate bo, take x=x_mean, y_cap=y_mean
b0=y_mean-b1*x_mean
print("The value of b0 is: ",b0)
```

```python
#so calculate y_cap
list5=[]
for i in df["x"]:
    l=b0+b1*i
    list5.append(l)
df["y_cap"]=list5
df
```

```python
#to calculate y_cap-y
list6=[]
for i in range(len(df["y_cap"])):
    list6.append(df["y_cap"][i]-df["y"][i])
df["y_cap-y"]=list6
df
```

```python
#y_cap-y**2
list7=[]
for i in df["y_cap-y"]:
    list7.append(i*i)
df["y_cap-y_sq"]=list7
df
```

```python
#to calculate std error
nume=df["y_cap-y_sq"].sum()
deno=(len(df))-2
std_error=math.sqrt(nume)/math.sqrt(deno)
print("The standard error is: ",std_error)
```

```python
if(std_error<1):
    print("Accept the model")
else:
    print("Reject the model")
```
